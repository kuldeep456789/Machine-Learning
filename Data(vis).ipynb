```python
#customizing line plot
x=np.arange(1,10,0.1)
y=2*x+5
plt.plot(x,y,color="r")
plt.title('LINE PLOT')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#sctter plot

x=[10,20,30,40,50,60,12]
a=[1,2,3,5,7,8,9]
b=[5,6,7,8,3,2,1]
plt.figure(figsize=(10,10))
plt.scatter(x,a,s=200)
plt.scatter(x,b,s=100,marker='3')
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('x')
plt.ylabel('a&b')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
data = [
    [1, 2, 3, 4, 5, 6, 7, 8],
    [2, 4, 6, 8, 10, 2, 6],
    [1, 2, 3, 4, 5, 6, 7, 8],
]

# Create violin plot
plt.violinplot(data, showmedians=True, showmeans=True)

# Grid and labels
plt.grid(True)
plt.title('Distribution of Data')
plt.xlabel('Data Set')
plt.ylabel('Value')

# Display plot
plt.show()
```

```python
#pie-chart
fruit=['apple','mango','orange','grapes']
quantity=[30,45,67,34]


plt.pie(quantity,labels=fruit,autopct='%0.1f%%',shadow=True ,explode=(0,0,0.2,0.1))
plt.show()
```
